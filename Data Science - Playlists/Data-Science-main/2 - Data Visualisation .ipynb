# Line Plot

```python
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y1)
plt.show()
```

```python
#Changing Line Color
plt.plot(x,y1,color='red')
plt.show()
```

```python
#Double Ploting
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.show()
```

```python
#To plot it in 2 different graphs :- 
plt.plot(x,y1,color='red')
plt.show()
plt.plot(x,y2,color='green')
plt.show()
```

```python
#Giving Titles
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='green')
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of Fruits over Time")
plt.show()
```

```python
#Giving Label
plt.plot(x,y1,color='red',label="Apple")
plt.plot(x,y2,color='green',label="Kiwi")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.title("Prices of Fruits over Time")
plt.show()
```

```python
#Dashed Line :-
plt.plot(x,y1,color='red',label="Apple")
plt.plot(x,y2,color='green',label="Kiwi",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.title("Prices of Fruits over Time")
plt.show()
```

```python
#Marking up the points :-
plt.plot(x,y1,color='red',label="Apple",marker='o')
plt.plot(x,y2,color='green',label="Kiwi",linestyle="dashed",marker='^')
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.title("Prices of Fruits over Time")
plt.show()
```

# Scatter Plot

```python
plt.scatter(x,y1,color='red',label="Apple",marker='o')
plt.scatter(x,y2,color='green',label="Kiwi",marker='^')
plt.xlabel("Time")
plt.ylabel("Prices")
plt.title("Scattered Plot - Prices of Fruits over Time")
plt.legend()
plt.show()
```

# Bar Graphs

```python
plt.bar([0,1,2],[10,20,15])
plt.show()
```

```python
plt.bar([0,1,2],[10,20,15]) #Current Year Prices
plt.bar([0,1,2],[20,10,12]) #Nexy Year Prices
plt.show()
```

```python
#Solving the Overlapping Problem

x_coordinates = np.array([0,1,2])
plt.bar(x_coordinates,[10,20,15]) #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12]) #Nexy Year Prices
plt.show()
```

```python
#Solving the Overlapping Problem

x_coordinates = np.array([0,1,2])
plt.bar(x_coordinates,[10,20,15],width=0.5) #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5) #Nexy Year Prices
plt.show()
```

```python
#Solving the Overlapping Problem

x_coordinates = np.array([0,1,2])*1.5
plt.bar(x_coordinates,[10,20,15],width=0.5) #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5) #Nexy Year Prices
plt.show()
```

```python
#Solving the Overlapping Problem

x_coordinates = np.array([0,1,2])*1.5
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year") #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5,label="Next Year") #Nexy Year Prices
plt.legend()
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*1.5
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platnium","Silver"]) #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5,label="Next Year") #Nexy Year Prices
plt.legend()
plt.show()
```

```python
x_coordinates = np.array([0,1,2])*1.5
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platnium","Silver"]) #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5,label="Next Year") #Nexy Year Prices
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")

x_coordinates = np.array([0,1,2])*1.5
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platnium","Silver"],color="green") #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5,label="Next Year",color="blue") #Nexy Year Prices
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Prices")
plt.legend()
plt.show()
```

```python
plt.style.use("seaborn")

x_coordinates = np.array([0,1,2])*1.5
plt.bar(x_coordinates,[10,20,15],width=0.5,label="Current Year",tick_label=["Gold","Platnium","Silver"]) #Current Year Prices
plt.bar(x_coordinates+0.5,[20,10,12],width=0.5,label="Next Year") #Nexy Year Prices
plt.title("Metal Price Comparision")
plt.xlabel("Metal")
plt.ylabel("Prices")
plt.ylim(0,30)
plt.xlim(-2,5)
plt.legend()
plt.show()
```

# Pie Charts

### To open Documentation\
`plt.pie?`

```python
subject = "Maths","Chemistry","English","Physics"
weightage = [20,10,5,15]

plt.pie(weightage,labels=subject)
plt.show()
```

```python
#To Highlight a subject seprately :-

subject = "Maths","Chemistry","English","Physics"
weightage = [20,10,5,15]

plt.pie(weightage,labels=subject,explode=(0,0,0,0.1))
plt.show()
```

```python
# To Check percentages :- 

subject = "Maths","Chemistry","English","Physics"
weightage = [20,10,5,15]

plt.pie(weightage,labels=subject,explode=(0,0,0,0.1),autopct='%1.1f&&')
plt.show()
```

```python
# To get Shadows :- 

subject = "Maths","Chemistry","English","Physics"
weightage = [20,10,5,15]

plt.pie(weightage,labels=subject,explode=(0,0,0,0.1),autopct='%1.1f&&',shadow=True)
plt.show()
```

```python
# To get Title :- 

subject = "Maths","Chemistry","English","Physics"
weightage = [20,10,5,15]
plt.title("Subjects")
plt.pie(weightage,labels=subject,explode=(0,0,0,0.1),autopct='%1.1f&&',shadow=True)
plt.show()
```

# Histogram

```python
xsn = np.random.randn(100)
print(xsn)
```

```python
xsn = np.random.randn(100)
sigma = 5
u = 60
x = xsn*sigma + u
print(x)
```

```python
#To round off

xsn = np.random.randn(100)
sigma = 8 #Standard Deviation
u = 70
x = np.round(xsn*sigma + u)
print(x)
```

```python
plt.hist(x)
plt.show()
```

```python
plt.hist(x)
plt.xlabel("Marks Range")
plt.ylabel("Prob/Freq Count of Students")
plt.show()
```

```python
xsn = np.random.randn(100)
sigma = 8 #Standard Deviation
u = 70
x = np.round(xsn*sigma + u)

x2 = np.round(xsn*5 + 40)

print(x,x2)
```

```python
plt.hist(x)
plt.hist(x2)
plt.xlabel("Marks Range")
plt.ylabel("Prob/Freq Count of Students")
plt.show()
```

```python
# Adding Opaquacity to see overlapping

plt.hist(x,alpha=0.8,label="Physics")
plt.hist(x2,alpha=0.8,label="Maths")
plt.xlabel("Marks Range")
plt.legend()
plt.ylabel("Prob/Freq Count of Students")
plt.show()
```
