```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Images/opencv21.webp')
```

```python
print(img.shape)
```

```python
plt.imshow(img)
```

### When ever open cv reads image, it read in BGR format, 

1st channel as B
2nd channel as G
3rd Channel as R

but we want to read it as RGB

### We need to convert BGR to RGB

```python
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#gray_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_image)
plt.show()
```

```python
plt.imshow(rgb_image)
plt.axis("off")
plt.show()
```

# Cropping Image

```python
plt.imshow(rgb_image)
plt.show()
```

```python
crop_image = rgb_image[20:250,50:225,:] #Slicing Columns, Slicing Rows, Slicing Channels
plt.imshow(crop_image)
plt.show()
```

```python
crop_image = rgb_image[20:250,50:225,:]
plt.imshow(crop_image)
plt.axis("off")
plt.show()
```
