## Data Visualisation Challenge

Given a Movies Dataset, your taskis is to make a visualisation that -
- Plots the Length of Movie Title Names on X-Axis
- Frequency Counts on Y-axis ie Number of Movies having 'x' characters in their title


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV File
df = pd.read_csv("movie_metadata.csv")
```

```python
df.head(n=10)
```

```python
# To find all the column names
df.columns
```

```python
# To get the specific column

titles = df.get('movie_title')
```

```python
print(titles)
```

```python
# Type Casting into List

titles = list(df.get('movie_title'))
```

```python
print(type(titles))
```

```python
# PRinting first five titles

print(titles[:5])
```

```python
freq_titles = {}

for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
#print(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie Data Visualiation Problem")
plt.show()
```
