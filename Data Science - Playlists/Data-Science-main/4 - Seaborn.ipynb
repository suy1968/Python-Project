# Seaborn is the SuperSet of the matplotlib
### With some advance statical data visualisation techniques

```python
import seaborn as sns
import numpy as np
```

[Seaborn Official Documentation](https://seaborn.pydata.org/examples/index.html)


### Check the gallery for all type of advance plotting graphs styles

```python
#Loading some random dataset using pandas

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
#Bar Graph

sns.barplot(x = 'sex' , y = 'total_bill' , data = tips)
```

```python
# Standard Deviation

sns.barplot(x = 'sex' , y = 'total_bill' , data = tips , estimator=np.std)
```

```python
# Count Plot

sns.countplot(x='sex',data=tips)
```

```python
# Count Plot on the other variables

sns.countplot(x='time',data=tips)
```

```python
# Box Plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
 sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
 sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
# Violin Plot

sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# Distribution Plot
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.displot(tips['total_bill'],kde=False)
```

```python
sns.displot(tips['total_bill'],kde=False,bins=40)
```

```python
sns.displot(tips['total_bill'],kde=False,bins=100)
```

```python
# KDE Plot
```

```python
sns.kdeplot(tips['total_bill'])
```

# Joint Plot

### For comparing two features or columns

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

# Pair Plot

### Plot pair wise relationship amosgst all features

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

# Pivot Table

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot=flights.pivot_table(index="month",columns="year",values="passengers")
```

```python
flights_pivot
```

```python
sns.heatmap(flights_pivot)
```

```python
sns.heatmap(flights_pivot,cmap='coolwarm')
```
